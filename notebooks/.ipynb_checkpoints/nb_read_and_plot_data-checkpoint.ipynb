```python
import os
import mne
import re

import numpy as np
import pandas as pd
import xarray as xr

from neo.rawio import ElanRawIO
from neo.io import ElanIO

import matplotlib.pyplot as plt

from mne.io import RawArray
from mne import create_info, Epochs

# %matplotlib widget
%load_ext autoreload
%autoreload 2
```

# Load the data

```python
root = '/envau/work/bagamore/brovelli.a/Data/Neurophy/SEEG_EBRAINS/Cohort I/seeg'
subject = 'HID-Sub-001'
eegfile = f"{subject}_MCSE_f50f150_ds8_sm0.eeg"
posfile = f"{subject}_MCSE_ds8.pos"

eegpath = os.path.join(root, subject, eegfile)
pospath = os.path.join(root, subject, posfile)

# sanity check
assert os.path.isdir(root)
assert os.path.isfile(eegpath)
assert os.path.isfile(pospath)

reader = ElanIO(eegpath, posfile=pospath)
reader.parse_header()
block = reader.read(lazy=False)  # get the first block
segment = block[0].segments         # get data from first (and only) segment
signals = segment[0].analogsignals[0]  # get first (multichannel) signal
data = signals.rescale('V').magnitude
unit = 'V'
```

# Read channel names + sampling rate

```python
sfreq = signals.sampling_rate.magnitude
```

```python
def clean_contacts(ch_names):
    ch_names = [n.split('.')[0].lower().strip().replace(' ', '').replace("'", "p") for n in ch_names]
    # add a 0 for single digit
    for n_i, n in enumerate(ch_names):
        if '-' in n: continue  # skip bipolar cleaning
        letter = re.findall(r'[a-z]*', n)[0]
        number = re.findall(r'\d+', n)[0]
        if len(number) == 1:
            number = '0' + number
        ch_names[n_i] = letter + number
    return ch_names
```

```python
# get the names from header
ch_names = reader.header['signal_channels']['name']
# clean contact names
ch_names = clean_contacts(ch_names)
```

# Read events

```python
(times_stamps, _, event_id) = reader.get_event_timestamps()
times_stamps, event_id = times_stamps.astype(int), event_id.astype(int)
events = np.c_[times_stamps, np.zeros((len(event_id),)), event_id].astype(int)
# events
```

# Load anatomy

```python
anat_root = '/envau/work/bagamore/brovelli.a/Data/Neurophy/SEEG_EBRAINS/Cohort I/anat'
path_anat = os.path.join(anat_root, subject, f"{subject}.csv")

assert os.path.isfile(path_anat)

# read the anatomy
anat = pd.read_csv(path_anat, skiprows=2, sep='\t')
# clean anat contacts
anat['contact'] = clean_contacts(list(anat['contact']))
# reindex according to channel names
anat = anat.set_index('contact').loc[ch_names].reset_index()
roi = list(anat['MarsAtlas'])
```

# MNE conversion
## Raw object + events

```python
info = create_info(ch_names, sfreq, ch_types='seeg')
raw = RawArray(data.T, info)
```

```python
epochs = Epochs(raw, events, event_id=dict(bottom=1, top=2), tmin=-1., tmax=3.)
```

# Xarray conversion

```python
da = xr.DataArray(epochs.get_data(), dims=('trials', 'roi', 'times'),
                  coords=(epochs.events[:, 2], roi, epochs.times))
conv = {1: 'T bottom', 2: 'T top'}
da['trials'] = [conv[k] for k in da['trials'].data]
da_gp = da.groupby('trials').mean('trials').groupby('roi').mean('roi')
```

```python
da_gp.sel(roi=['L_Mdl', 'L_Mdm', 'L_Mv']).plot.line(x='times', hue='trials', col='roi')
```

# Plotly plotting

```python
import plotly.express as px
```
